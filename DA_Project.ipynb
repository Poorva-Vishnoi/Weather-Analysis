```python
def plot_vars(data,x,y,month):
    fig,ax=plt.subplots(figsize=(20,10))
    plt.grid(True)
    plt.xlabel('Date', fontsize=20)
    plt.ylabel('Temperature', fontsize=20)
    sns.lineplot(
        ax=ax,
        data=df,
        x=x,
        y=y,
        hue=month,
    palette='dark',
    linestyle='solid',
    linewidth='2'
    )
    plt.show()
```

```python
plot_vars(data=df, x='Summary',y='Date', month='Month')
```

```python
plot_vars(data=df,x='Date',y='Temperature',month='Month')
```

```python
def plot_vars(data,x,y,month):
    fig,ax=plt.subplots(figsize=(20,10))
    plt.title("Humidity")
    plt.xlabel('Date', fontsize=20)
    plt.ylabel('Humidity', fontsize=20)
    plt.grid(True)
    sns.lineplot(
        ax=ax,
        data=df,
        x=x,
        y=y,
        hue=month,
    palette='dark',
    linestyle='dotted',
    linewidth='2'
    )
    plt.show()
plot_vars(data=df,x='Date',y='Humidity',month='Month')
```

```python
sns.relplot(x="Humidity",y="Temperature",data=df,col="Month",col_wrap=3)
```

```python
def rel_temp(data,x,y):
    fig,ax=plt.subplots(figsize=(20,10))
    plt.title("Summary of sky")
    plt.xlabel('Date', fontsize=20)
    plt.ylabel('WindSpeed', fontsize=20)
    sns.scatterplot(
        ax=ax,
        data=df,
        x=x,
        y=y,
    size='Summary',
    hue='Summary',
    palette='bright',
    style='Summary',
    sizes=(20, 200), legend="full"
    
    )
    plt.show()
rel_temp(data=df,x='Date',y='WindSpeed')
```

```python
def rel_temp(data,x,y):
    fig,ax=plt.subplots(figsize=(20,10))
    plt.title("Pressure")
    plt.xlabel('Date', fontsize=20)
    plt.ylabel('Pressure', fontsize=20)
    sns.scatterplot(
        ax=ax,
        data=df,
        x=x,
        y=y,
    size='Precipitation',
    hue='Precipitation',
    palette='flare',
    style='Precipitation',
    sizes=(200, 200), legend="full"
    )
    plt.show()
```

```python
rel_temp(data=df,x='Date',y='Pressure')
```
